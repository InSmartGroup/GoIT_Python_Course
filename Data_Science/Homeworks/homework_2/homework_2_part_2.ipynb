```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the csv file
df = pd.read_csv("2017_jun_final - 2017_jun_final.csv",
                 sep=",",
                 skip_blank_lines=False)

df.head(10)
```

```python
# check the shape
print(f"DataFrame's shape: {df.shape}")  # 8704 rows by 25 columns
```

```python
# check column data types
df.dtypes
```

```python
# check how many missing values we have
df.isna().sum()
```

```python
# remove all the columns with missing values except for the "Мова програмування" column
# first, let's save the needed Series
df_needed = df["Мова програмування"]
```

```python
# second, let's drop all the columns with missing values
df.dropna(axis=1, inplace=True)
```

```python
# finally, let's add the needed column to our DF
df["Мова програмування"] = df_needed
```

```python
# check how many missing values we have
df.isna().sum()  # now it's zeros everywhere, except for the needed column
```

```python
# remove all the rows with missing values
df.dropna(axis=0, inplace=True)
```

```python
# check the shape
print(f"DataFrame's shape: {df.shape}")  # 4854 rows by 20 columns
```

```python
# create a new sheet which contains only the rows with specialties that have Python as a programming language
python_data = df[df["Мова програмування"] == "Python"]
```

```python
# check the shape of the new DF
print(f"DataFrame's shape: {python_data.shape}")  # 207 rows by 20 columns

python_data.head(10)
```

```python
# create a new DF by column "Посада" and perform data aggregation using the agg() method, and calculate minimum and maximum values
python_data_aggregated = python_data.agg({"Зарплата на місяць": ["min", "max"]})

python_data_aggregated
```

```python
# write a function that will return an average salary per month
def fill_avg_salary(row):
    return python_data.loc[python_data["Посада"] == row["Посада"], ["Зарплата на місяць"]].mean()

# add these values to a new column
python_data["avg"] = python_data.apply(fill_avg_salary, axis=1)

python_data["avg"].describe()
```

```python
# save the new DF as a csv file
python_data.to_csv("python_programmers.csv",
                   sep=",",
                   header=True,
                   index=False)
```
